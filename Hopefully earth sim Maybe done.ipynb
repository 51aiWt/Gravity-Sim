```python
#1px=0.005AU   1px=7.48*10^8m
import pygame
import math 
from sys import exit
pygame.init()
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Gravitational Simulation")
clock = pygame.time.Clock()
#Initial parameter
x_1=1.496 *10**11 #meters x postion of the earth from the sun
y_1=0 #meter the y postion of the earth
v_x_1=0 #m/s the x velocity of the earth
T=3.1553280 * 10**7 #seconds 365.24 days Period of earth around the sun
w=(2 * math.pi)/T #angular velocity of earth around the sun rad/s
v_y_1=w*x_1 #m/s the y velocity of the earth
m_0=1.988*10**30 #kg mass of the sun
m_1=5.972 *10**24 #kg mass of the earth
r_10=(x_1**2 + y_1**2)**0.5 #meters distance between the sun and the earth
G=6.674 * 10**-11 #Nm^2/kg^2 newton's gravitational constant 
t=0 #time elapsed in seconds (s)
time_scaled=10**7 #in seconds (s)
path_points=[] # a list to store the path points of the earth on a x,y coordinate system
stopped=False  #store the false value to dictate when the simulation stops running
# Sun representation
sun_surface = pygame.Surface((50, 50), pygame.SRCALPHA) 
pygame.draw.circle(sun_surface, (255, 255, 0), (25, 25), 25)
#Earth representation
earth_surface=pygame.Surface((20,20),pygame.SRCALPHA)
pygame.draw.circle(earth_surface,(0,0,255),(10,10),10)
#while loop to keep the simulation running and to update the position of the earth using physics
while True:
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            pygame.quit()
            exit()

    if not stopped:
        dt = (clock.tick(60) / 1000) *time_scaled #This takes a small timestep of 10^-3 s or 1 millsecond and converts it to well seconds or 10^-3 s. It also keeps the sim running at 60fps. In (s)
        t+= dt #This takes the variable t and adds our dt cummulatively. Additionally the varialbe time_scale is used to up scale the time of the system and thus the movement in seconds (s).


        




        # scalar Î”v magnitude from gravity
        delta_v_one = (G * m_0 / (r_10**2)) * dt # gives the velocity of earth m/s based on forced applied to earth

        # angle of the position vector using your acos branch, physical y_1 used and angel units as in degress.
        if y_1 >= 0:
            theta_1 = math.acos(x_1 / r_10) #Both angles give the angle of the earth in relation to the postion and radius of the earth (right triangle relationship)
        else:
            theta_1 = -math.acos(x_1 / r_10)

        # gravity points TOWARD the origin (negate components ?). The velocity has m/s.
        delta_vx = -delta_v_one * math.cos(theta_1) #both gives the separate velocity component on a xy coordinate plane since the main velocity component 
        delta_vy = -delta_v_one * math.sin(theta_1)

        # new velocities (temporaries) m/s
        v_primex = v_x_1 + delta_vx
        v_primey = v_y_1 + delta_vy

        # new positions (temporaries) meter in x and y components
        delta_x = v_primex * dt
        delta_y = v_primey * dt
        x_prime = x_1 + delta_x
        y_prime = y_1 + delta_y

        # STORE temporaries back into state so motion persists. Switches new variable value to old variables updating the old variables
        v_x_1 = v_primex #m/s
        v_y_1 = v_primey#m/s 
        x_1 = x_prime #meters
        y_1 = y_prime #meters



        # update distance for next step
        r_10=(x_1**2 + y_1**2)**0.5 #meters of the new updated radius






        # meters -> pixels
        x = (x_1 / (7.48*10**8)) + 400
        y = (y_1 / (7.48*10**8)) + 300

        path_points.append((int(x), int(y)))

    screen.fill((0, 0, 0))
    for point in path_points:
        pygame.draw.circle(screen, (0, 255, 255), point, 2)
    screen.blit(earth_surface, (x-10, y-10))
    screen.blit(sun_surface, (400 - 25, 300 - 25))
    pygame.display.update()
    clock.tick(60)


        
```
